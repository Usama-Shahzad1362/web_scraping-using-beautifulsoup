```python
# download the paklap_price file and save it in the local drive 

import requests 
from fake_useragent import UserAgent

user_agent = UserAgent()
header = {'user_agent':user_agent.chrome}
url = 'https://www.paklap.pk/laptops-prices/hp-laptops.html'
res = requests.get(url,headers = header)
res.raise_for_status()
scrape_File = open('paklap_price.txt', 'wb')
for chunk in res.iter_content(100000):
    scrape_File.write(chunk)   
scrape_File.close()
```

```python
# scrape the price,link and description of the H-P laptops and strore it in the csv file using pandas

from bs4 import BeautifulSoup
import pandas as pd

def readfile():
    file = open("paklap_price.txt")
    data = file.read()
    return data
    file.close()

soup = BeautifulSoup(readfile(),"lxml")

all_laptops = soup.find_all("li",class_="item product product-item")
short_desc = [soup.find("strong",class_="product name product-item-name").get_text().strip() for data in all_laptops]
all_links = [soup.find("a",class_="product-item-link")["href"] for link in all_laptops]
price = [soup.find("span",class_="price").get_text().strip() for link in all_laptops]

laptops_data = pd.DataFrame({"Short_Description":short_desc,
                            "Links":all_links,
                            "Price":price,})
print(laptops_data)

laptops_data.to_csv("laptops_data.csv")
```
